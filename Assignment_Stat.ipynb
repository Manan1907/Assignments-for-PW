```python
#9.	Write a Python script to calculate covariance between two datasets

def calculate_covariance(data1, data2):

    if len(data1) != len(data2) == 0:
        raise ValueError("Datasets must have the same length.")

    n = len(data1)
    mean_data1 = np.mean(data1)
    mean_data2 = np.mean(data2)

    covariance = np.sum((data1 - mean_data1) * (data2 - mean_data2)) / (n - 1)
    return covariance

if __name__ == '__main__':
    dataset1 = [1, 2, 3, 4, 5, 6]
    dataset2 = [9, 8, 7, 6, 5, 4]

    try:
        covariance = calculate_covariance(dataset1, dataset2)
        print(f"The covariance between the two datasets is: {covariance}")
    except ValueError as e:
        print(f"Error: {e}")
```

```python
#10.	Write a Python script to calculate the correlation coefficient between two datasets


def calculate_correlation(data3, data4):
    if len(data3) != len(data4) == 0:
        raise ValueError("Datasets must have the same length.")

    try:
        corr, _ = pearsonr(data3, data4)
        return corr
    except Exception as e:
        print(f"Error calculating correlation: {e}")
        return None


if __name__ == "__main__":
    data3 = [1, 2, 3, 4, 5, 6]
    data4 = [9, 8, 7, 6, 5, 4]

    correlation = calculate_correlation(data3, data4)

    if correlation is not None:
        print(f"The correlation coefficient between the two datasets is: {correlation}")

    else:
        print("Error calculating correlation.")
```

```python
#12.	Implement and compare simple random sampling and systematic sampling


#A. Simple random sampling

def simple_random_sampling(population, sample_size):
  if sample_size > len(population):
    raise ValueError("Sample size cannot exceed population size.")
  return random.sample(population, sample_size)

population = list(range(1, 101))  # Population of 100 elements
sample_size = 10
random_sample = simple_random_sampling(population, sample_size)
print("A. Simple Random Sample:", random_sample)


#B. Systamatic sampling
def systematic_sampling(population, sample_size):

    population_size = len(population)
    if sample_size > population_size:
        raise ValueError("Sample size cannot exceed population size.")
    interval = population_size // sample_size
    start = random.randint(0, interval - 1)
    sample = [population[i] for i in range(start, population_size, interval)][:sample_size]
    return sample

population = list(range(1, 101))
sample_size = 20
systematic_sample = systematic_sampling(population, sample_size)
print("\nB. Systematic Sample:", systematic_sample)
```

```python
#22.	Generate a dataset and implement both variance and standard deviation computations


# Generate a sample dataset

def generate_dataset(size=50, seed=42):
    np.random.seed(seed)
    return np.random.rand(size)

# Calculate the mean of a dataset
def calculate_mean(data_22):
    return sum(data_22) / len(data_22)

# Calculate the variance of a dataset
def calculate_variance(data_22):
    n = len(data_22)
    if n < 2:
        return 0 # Variance is not defined for datasets with less than 2 elements
    mean = calculate_mean(data_22)
    squared_differences = [(x - mean) ** 2 for x in data_22]
    return sum(squared_differences) / (n - 1) # Using n-1 for sample variance


# Calculate the standard deviation of a dataset
def calculate_standard_deviation(data_22):
    variance = calculate_variance(data_22)
    return math.sqrt(variance)

# Generate dataset
dataset = generate_dataset()

# Calculate and print statistics
mean = calculate_mean(dataset)
variance = calculate_variance(dataset)
std_dev = calculate_standard_deviation(dataset)

print(f"1. Dataset:\n{dataset}")
print(f"\n2. Mean: {mean:.2f}")
print(f"\n3. Variance: {variance:.2f}")
print(f"\n4. Standard Deviation: {std_dev:.2f}")
```

```python
#23.	Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn


# Generate sample data

np.random.seed(42)
data_normal = np.random.normal(0, 1, 1000) # Normal distribution
data_skewed_pos = np.random.exponential(1, 1000) # Positive skew
data_skewed_neg = -np.random.exponential(1, 1000) # Negative skew
data_high_kurtosis = np.random.laplace(0, 1, 1000) # High kurtosis

data_list23 = [data_normal, data_skewed_pos, data_skewed_neg, data_high_kurtosis]
titles = ["Normal Distribution", "Positive Skew", "Negative Skew", "High Kurtosis"]

for i, data_23 in enumerate(data_list23):
    # Calculate Skewness and Kurtosis
    skewness = skew(data_23)
    kurt = kurtosis(data_23, fisher=False)

    # Create a figure and axes
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Plotting the Distribution
    sns.histplot(data_23, kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(titles[i])
    axes[0].axvline(np.mean(data_23), color='green', linestyle='--') # Mean
    axes[0].axvline(np.median(data_23), color='red', linestyle='--') # Median
    axes[0].text(0.97, 0.97, f"Skew: {skewness:.2f}\nKurt: {kurt:.2f}", transform=axes[0].transAxes,
                fontsize=10, verticalalignment='top', horizontalalignment='right',
                backgroundcolor='white', color='xkcd:poo brown')

    # Plotting the Box Plot
    sns.boxplot(x=data_23, ax=axes[1], color='lightcoral')
    axes[1].set_title("Box Plot")

    plt.tight_layout()
    plt.show()
```
